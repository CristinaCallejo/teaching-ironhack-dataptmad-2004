# Error Evaluation

## 1. Regression

### 1.1 Data Loading

```python
from sklearn.datasets import load_boston

data = load_boston()

data
```

### 1.2 Modeling

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data.data, data.target)

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor()
rf.fit(X_train, y_train)

y_pred = rf.predict(X_test)

y_pred[:10]
```

```python
y_test[:10]
```

### 1.3 Metrics

#### 1.3.1 MSE (Mean Square Error)

$$MSE = \frac{1}{n}\sum_{i=1}^{n}(y_i-\hat{y}_i)^{2}$$

```python
from sklearn.metrics import mean_squared_error

mean_squared_error(y_test, y_pred)
```

#### 1.3.2 RMSE (Root Mean Square Error)

$$RMSE = \sqrt{\frac{1}{n}\sum_{i=1}^{n}(y_i-\hat{y}_i)^{2}}$$

```python
mean_squared_error(y_test, y_pred)**0.5
```

#### 1.3.3 MAE (Mean Absolute Error)

$$MAE = \frac{1}{n}\sum_{i=1}^{n}|y_i-\hat{y}_i|$$

```python
from sklearn.metrics import mean_absolute_error

mean_absolute_error(y_test, y_pred)
```

#### 1.3.4 R2

$$R2 = 1 - \frac{\sum_{i=1}^{n}(y_i-\hat{y}_i)^{2}}{\sum_{i=1}^{n}(y_i-\bar{y})^{2}}$$

```python
from sklearn.metrics import r2_score

r2_score(y_test, y_pred)
```

### 1.4 Evaluation

```python
rf.score(X_test, y_test)
```

## 2. Classification

### 2.1 Data Loading

```python
from sklearn.datasets import load_breast_cancer

data = load_breast_cancer()

data
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data.data, data.target)

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier()
rf.fit(X_train, y_train)

y_pred = rf.predict(X_test)
y_prob = rf.predict_proba(X_test)

print(y_pred[:10])
print(y_prob[:10])
```

### 2.3 Metrics

#### 2.3.0 Definitions

+ TP = True Positives (predict 1 when 1)
+ TN = True Negatives (predict 0 when 0)
+ FP = False Positive (predict 1 when 0)
+ FN = False Negative (predict 0 when 1)

+ Accuracy = (TP+TN)/(TP+TN+FP+FN)
+ Precision = TP/(TP+FP)
+ Recall = TP/(TP+FN)

#### 2.3.1 Accuracy

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test, y_pred)
```

#### 2.3.2 Precission

```python
from sklearn.metrics import precision_score

precision_score(y_test, y_pred)
```

#### 2.3.3 Recall

```python
from sklearn.metrics import recall_score

recall_score(y_test, y_pred)
```

#### 2.3.4 Confussion Matrix

```python
from sklearn.metrics import confusion_matrix

confusion_matrix(y_test, y_pred)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

ax = sns.heatmap(confusion_matrix(y_test, y_pred), annot=True)

b, t = ax.get_ylim()
ax.set_ylim(b + 0.5, t - 0.5)

plt.title('Confussion Matrix')
plt.ylabel('Ground Truth')
plt.xlabel('Prediction')
plt.show();
```

#### 2.3.4 ROC Curve

+ TPR = True Positive Rate = TP/(TP+FN)
+ FPR = False Positive Rate = FP/(TN+FP)

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score

roc_curve(y_test, y_pred)
```

```python
# wrong plot (think about threshold variation)
with plt.xkcd():
    fpr, tpr, thresholds = roc_curve(y_test, y_pred)
    a = roc_auc_score(y_test, y_pred)
    plt.plot(fpr, tpr)
    plt.plot(fpr, fpr, 'r--')
    plt.xlabel('FPR')
    plt.ylabel('TPR')
    plt.title(f'Prob ROC Curve - AUC={a:.3f}')
    plt.show();
```

### 2.4 Evaluation

```python
rf.score(X_test, y_test)
```
